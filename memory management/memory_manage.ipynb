```python
class MyObject:
    def __init__(self,name):
        self.name=name
        print(f"object {self.name} created")

    def __del__(self):
        print(f"object {self.name} deleted")

obj1=MyObject("obj1")
obj2=MyObject("obj2")

obj1.ref=obj2
obj2.ref=obj1

del obj1
del obj2

gc.collect()
```

```python
def generators_number(n):
    for i in range(n):
        yield(i)

for num in generators_number(100000):
    print(num)
    if num>10:
        break
```
