```python
students={
    "student1":{"name":"omkar","age":22},
    "student2":{"name":"vishal","age":24},
    "student3":{"name":"atharva","age":23}
}
```

```python
print(students["student2"]["name"])
```

```python
students.items()
```

```python
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```
