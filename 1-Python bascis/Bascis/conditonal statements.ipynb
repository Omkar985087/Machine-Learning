```python
num1=float(input("enter the number 1"))
num2=float(input("enter the number 2"))
operation=input("enter the operation(+,-,/,*):")

if operation=='+':
    result=num1+num2
elif operation=='-':
    result=num1-num2
elif operation=='/':
    result=num1/num2
else:
    result=num1*num2

print(result)
```
