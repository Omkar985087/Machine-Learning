```python
def print_numbers(*args):
    for i in args:
        print(i)
```

```python
print_numbers(1,2,3,4,5,6)
```

```python
def print_info(**kargs):
    for key,value in kargs.items():
        print(f"{key}:{value}")
```

```python
print_info(name="omkar",age="34")
```

```python
def convert_temp(temp,unit):
    if unit=='C':
        return temp*2
    elif unit=='F':
        return temp/2
    else:
        return None

result=convert_temp(42,'C')
print(result)
```

```python
def calculate(cart):
    result=0
    for i in cart:
        result+=i['price']*i['quantity']
    return result

cart=[
    {'name':'apple','price':4,'quantity':4},
    {'name':'banana','price':5,'quantity':6},
    {'name':'mango','price':6,'quantity':8}
]

result=calculate(cart)
print(result)
```
