```python
def convert_temp(temp,unit):
    if unit=="C":
        return temp*(9/5)+32
    elif unit=="F":
        return (temp-32)*5/9
    else:
        return None
    
print(convert_temp(25,'C'))
print(convert_temp(55,'F'))
```

```python
def strong_password(password):
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#%&*()' for char in password):
        return False
    return True

print(strong_password("Omkar@9850"))
print(strong_password("omkar"))
    
```

```python
def caculate(cart):
    total_cost=0
    for i in cart:
        total_cost+=i['price']*i['quantity']
    
    return total_cost


cart=[
    {'name':'apple','price':4,'quantity':4},
    {'name':'banana','price':5,'quantity':6},
    {'name':'mango','price':6,'quantity':8}
]

total_cost=caculate(cart)
print(total_cost)
```

```python
def is_palindrome(s):
    s=s.lower().replace(" ","")
    return s==s[::-1]

print(is_palindrome("A man a plan a canal Panama"))
```
