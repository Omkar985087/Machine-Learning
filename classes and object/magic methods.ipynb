```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    
    def __str__(self):
        return f"{self.name},{self.age} years old"
    
    def __repr__(self):
        return f"person(name={self.name},age={self.age})"

person=Person("omkar",22)
print(person)
print(str(person))
print(repr(person))
```
