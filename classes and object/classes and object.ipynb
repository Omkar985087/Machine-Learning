```python
class bankaccount:
    def __init__(self,owner,balance=0):
        self.owner=owner
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print(f"{amount} is depositd.new balance is {self.balance}")

    def withdraw(self,amount):
        if amount>self.balance:
            print("you cannot withdraw")
        else:
            self.balance-=amount
            print(f"this {amount} is withdrawn.new balance is {self.balance}")
    
    def get_balance(self):
        return self.balance

account=bankaccount("omkar",5000)
print(account.balance)
```

```python
account.deposit(100)
```

```python
account.withdraw(300)
```
