```python
import sqlite3
```

```python
connection=sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
cursor.execute('''
Create Table If Not Exists employee(
    id Interger Primary Key,
    name ext Not Null,
    age Integer,
    department text
    )
''')

connection.commit()
```

```python
cursor.execute('''
Select * from employee
''')
```

```python
cursor.execute('''
Insert Into employee(name,age,department)
                values('omkar',22,'data scientist')
''')

cursor.execute('''
Insert Into employee(name,age,department)
                values('atharva',24,'doctor')
''')

cursor.execute('''
Insert Into employee(name,age,department)
                values('vishal',26,'drug dealer')
''')

connection.commit()
```

```python
cursor.execute('Select * from employee')
rows=cursor.fetchall()

for row in rows:
    print(row)
```
